```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading Data into DF
custom_separator = ', '
filename0 = 'stride_O0.csv'
filename2 = 'stride_O2.csv'
df0 = pd.read_csv(filename0, sep=custom_separator, engine='python')
df2 = pd.read_csv(filename2, sep=custom_separator, engine='python')
```

```python
df0.head()
```

```python
df2.head()
```

```python
df2.columns
```

```python
stride_O0 = df0['stride ']
time_O0 = df0['time (msec)']

stride_O2 = df2['stride ']
time_O2 = df2['time (msec)']
```

```python
plt.figure(figsize=(10, 6))

plt.plot(stride_O0, time_O0, label='-O0', marker='o')
plt.plot(stride_O2, time_O2, label='-O2', marker='x')

plt.xlabel('Stride')
plt.ylabel('CPU Time (msec)')
plt.title('CPU Time vs. Stride')
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Extraire les valeurs nécessaires pour le tracé

bandwidth_O0 = df0['rate (MB/s)']

bandwidth_O2 = df2['rate (MB/s)']

# Créer un graphique
plt.figure(figsize=(10, 6))

plt.plot(stride_O0, bandwidth_O0, label='-O0', marker='o')
plt.plot(stride_O2, bandwidth_O2, label='-O2', marker='x')

plt.xlabel('Stride')
plt.ylabel('bandwidth (MB/s)')
plt.title('Bandwidth vs. Stride')
plt.grid(True)
plt.legend()

plt.show()
```
